```python
from diffusers import StableDiffusionPipeline
import torch

# Đường dẫn cụ thể tới concept của bạn trên Hugging Face
concept_token_path = "sd-concepts-library/irit2"

# Tải pipeline Stable Diffusion
model_id = "SG161222/Realistic_Vision_V6.0_B1_noVAE"  # Model Stable Diffusion cơ bản
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# Load concept token vào pipeline
pipe.load_textual_inversion(concept_token_path, token="irit1")  # Tên token từ README của bạn

# Sử dụng CUDA nếu có GPU
pipe.to("cuda")
```

```python
pos = "a photo of a <irit1> woman holding flowers, white background, Best Quality."
image = pipe(prompt=pos, num_inference_steps=30, guidance_scale=8).images[0]
image
```
