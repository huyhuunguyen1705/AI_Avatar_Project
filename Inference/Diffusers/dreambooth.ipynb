```python
from diffusers import StableDiffusionPipeline

# Tải mô hình DreamBooth từ Hugging Face
model_id = "sd-dreambooth-library/itay11"  # Thay thế bằng tên mô hình của bạn

# Tải pipeline Stable Diffusion đã fine-tuned bằng DreamBooth
pipe = StableDiffusionPipeline.from_pretrained(model_id)
```

```python
prompt = "a photo of <itay11> man "  # Ví dụ mô tả ảnh bạn muốn tạo
image = pipe(prompt).images[0]
```
